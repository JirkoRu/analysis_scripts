### Call R

```python
# Call R
from rpy2.robjects.packages import importr
from rpy2.robjects import IntVector, FloatVector, FactorVector, globalenv
lme4 = importr('lme4')
base = importr('base')
stats = importr('stats')
broom = importr('broom')
utils = importr('utils')
ez = importr('ez')
```

### Load data

```python
dname = os.getcwd() + '/../../data/'
DF = pd.read_csv(dname + 'full_data.csv')
```

```python
pd.set_option('display.max_columns', None)
DF.head(7)
```

### Effect of condition and trial index

```python
globalenv['correct'] = FloatVector(DF['response_accuracy'])
globalenv['suj'] = FactorVector(DF['id'])
globalenv['bloc_type'] = FactorVector(DF['bloc_type'])
globalenv['trial_index'] = FloatVector(DF['trial_index'])
```

```python
formula = 'correct ~ bloc_type + trial_index + (1|suj)'
m = lme4.glmer(formula, family='binomial')
```

```python
print(broom.tidy_gmm(m))
```
