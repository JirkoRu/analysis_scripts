```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
import pandas as pd
from scipy.stats import pearsonr
from sklearn.metrics import pairwise_distances
```

### Load data

```python
dname = os.getcwd() + '/../../data/'
```

```python
CONFUSION_MATRIX_MODELS = np.load(dname + 'CONFUSION_MATRIX_MODELS.npy')
CONFUSION_MATRIX_PARTICIPANTS = np.load(dname + 'CONFUSION_MATRIX_PARTICIPANTS.npy')
DIMS = np.load(dname + 'DIMS.npy', allow_pickle=True)
MODALITY = np.load(dname + 'MODALITY.npy', allow_pickle=True)
```

```python
modalities = np.array(['audio', 'visual', 'spatial'])
modalities_plot = np.array(['Auditory', 'Visual', 'Spatial (visual)'])
n_modalities = np.size(modalities)
```

### Global parameters

```python
model_names = np.array([
    'No learn.', 
    'No gen.', 
    '1Dx$\\bigstar$', 
    '1Dx$\\bigstar$x', 
    '1Dx$\\bigstar$y', 
    '1Dx', 
    '1Dy$\\bigstar$', 
    '1Dy$\\bigstar$x', 
    '1Dy$\\bigstar$y', 
    '1Dy', 
    '2D$\\bigstar$', 
    '2D'
])
```

```python
n_models, n_trial_types, n_labels, _ = CONFUSION_MATRIX_MODELS.shape
n_sujs = CONFUSION_MATRIX_PARTICIPANTS.shape[0]
```

### Fit model to participants data

```python
CONFUSION_MATRIX_MODELS.shape
```

```python
CONFUSION_MATRIX_PARTICIPANTS.shape
```

```python
CORRELATION_VALUES = np.full((n_sujs, n_models), np.nan)
DISTANCE_VALUES = np.full((n_sujs, n_models), np.nan)

# Loop across participant
for i in range(n_sujs):
    for j in range(n_models):

        # Get values
        x = CONFUSION_MATRIX_MODELS[j].flatten()
        y = CONFUSION_MATRIX_PARTICIPANTS[i].flatten()

        # Compute distance value
        DISTANCE_VALUES[i, j] = pairwise_distances(x[None, :], y[None, :], metric='euclidean')[0, 0]

        # Compute correlation
        if np.unique(x).size == 1 or np.unique(y).size == 1: continue
        CORRELATION_VALUES[i, j] = pearsonr(x, y)[0]
```

```python
DISTANCE_VALUES.shape
```

```python
# Choose model with the minimal distance 
best_fitting_model_idx = np.argmin(DISTANCE_VALUES, axis=1)
best_fitting_model = model_names[best_fitting_model_idx]
best_fitting_model.shape
```

```python
# Choose model with the maximal correlation
CORRELATION_VALUES[np.isnan(CORRELATION_VALUES)] = 0.4
best_fitting_model_idx = np.argmax(CORRELATION_VALUES, axis=1)
best_fitting_model = model_names[best_fitting_model_idx]
best_fitting_model.shape
```

### Plot fit

```python
# Re-order model_names 
model_names_plot = np.array([
    'No learn.',
    'No gen.', 
    '1Dx', 
    '1Dy', 
    '1Dx$\\bigstar$', 
    '1Dy$\\bigstar$', 
    '1Dx$\\bigstar$x', 
    '1Dy$\\bigstar$y', 
    '1Dx$\\bigstar$y', 
    '1Dy$\\bigstar$x', 
    '2D',
    '2D$\\bigstar$', 
])
```

```python
fig, axs = plt.subplots(n_modalities, 1, figsize=(9, 9), dpi=300, facecolor='w')

for i, (ax, modality, modality_plot) in enumerate(zip(axs, modalities, modalities_plot)):
    
    # Select subjects for this modality
    idx = MODALITY == modality
    
    # Plot bar 
    for j, model_name in enumerate(model_names_plot):
        ax.bar(model_name, np.sum(best_fitting_model[idx] == model_name))

    for x in np.array([0.5, 1.5, 3.5, 7.5, 9.5, 10.5]):
        ax.axvline(x, -10, 10, ls='--', color='k', alpha=0.3)

    ax.set_title(modality_plot, fontweight='bold')
    ax.set_xlabel('Model')
    ax.set_ylabel('# Participants')
    ax.set_ylim(0, 24)
    ax.set_yticks(range(0, 25, 2))

plt.tight_layout()
fig.savefig('figures/fig_3.png')
```
