```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import matplotlib

from sklearn.metrics import confusion_matrix
```

### Import files

```python
dname = os.getcwd() + '/../../data/'
dname
```

```python
modalities = ['audio', 'visual', 'spatial'] # ['temp'] #
```

```python
DF = pd.DataFrame()
for modality in modalities:
    
    # Search for file names
    fnames = glob.glob(dname + 'raw/' + modality + '/*.txt')

    for fname in fnames:

        # Read CSV
        _df = pd.read_csv(fname)

        # Add condition columun
        _df['modality'] = modality
        
        # Append to the big dataFrame
        DF = DF.append(_df, ignore_index=True)
```

### Retreive parameters

```python
sujs = DF['id'].unique()
n_sujs = sujs.size
n_sujs
```

```python
n_trials_per_suj = DF.loc[DF['id'] == sujs[-1]].shape[0]
n_trials_per_suj
```

```python
n_blocs = DF['bloc_index'].unique().size
n_blocs
```

```python
n_trials_training = int(DF.loc[DF['bloc_type'] == 'training', 'bloc_n_trials'].unique()[0])
n_trials_generalization = int(DF.loc[DF['bloc_type'] == 'generalization', 'bloc_n_trials'].unique()[0])
n_trials = np.max([n_trials_training, n_trials_generalization])
n_trials
```

### Put data in numpy arrays

```python
def binatodeci(binary):
    return sum(val*(2**idx) for idx, val in enumerate(reversed(binary)))

def convertTypeString(x):
    
    # Get binary list identifier
    bin_id = np.array([
        x['rot_0'], 
        x['rot_pi'], 
        x['rot_pi_2'], 
        x['rot_minus_pi_2'], 
        x['mirror_x'], 
        x['mirror_y'], 
        x['extra']
    ])
    
    # Convert binary list to number
    return binatodeci(bin_id)
```

```python
CORRECT_BLOC = np.full((n_sujs, n_blocs, n_trials), np.nan)
RT_BLOC = np.full((n_sujs, n_blocs, n_trials), np.nan)
RESPONSE_CATEGORY_BLOC = np.full((n_sujs, n_blocs, n_trials), np.nan, dtype=object)
CORRECT_CATEGORY_BLOC = np.full((n_sujs, n_blocs, n_trials), np.nan, dtype=object)
TRIAL_TYPE = np.full((n_sujs, n_blocs, n_trials), np.nan)

# Loop across participant
for i, suj in enumerate(sujs):
    for j in range(n_blocs):
        
        # Select relevant data
        x = DF.loc[(DF['id'] == suj) & (DF['bloc_index'] == j)]
        
        # Get response data (accuracy + rt)
        CORRECT_BLOC[i, j, :x.shape[0]] = x['response_accuracy']
        RT_BLOC[i, j, :x.shape[0]] = x['rt']

        # Category of the response 
        RESPONSE_CATEGORY_BLOC[i, j, :x.shape[0]] = x['response_category']

        # Correct category
        CORRECT_CATEGORY_BLOC[i, j, :x.shape[0]] = x['correct_sequence_category']
        
        # Trial type
        for k in range(x.shape[0]):
            TRIAL_TYPE[i, j, k] = convertTypeString(x.iloc[k])
```

### Information about blocs in arrays

```python
BLOC_TYPE = np.full((n_sujs, n_blocs, n_trials_training), np.nan, dtype=object)
BLOC_CENTER_TRIAL = np.full((n_sujs, n_blocs), np.nan, dtype=object)
BLOC_STARTING_TRIAL = np.full((n_sujs, n_blocs), np.nan, dtype=object)
for i, suj in enumerate(sujs):
    for j in range(n_blocs):
        
        # Bloc type
        if DF.loc[(DF['id'] == suj) & (DF['bloc_index'] == j), 'bloc_type'].unique().size == 0: continue
        BLOC_TYPE[i, j, :] = DF.loc[(DF['id'] == suj) & (DF['bloc_index'] == j), 'bloc_type'].unique()[0]
        
        # Center trial index
        previous_index = np.sum(BLOC_TYPE[i, :j, 0] == 'training') * n_trials_training + \
                         np.sum(BLOC_TYPE[i, :j, 0] == 'generalization') * n_trials_generalization
        this_index = n_trials_training / 2 if BLOC_TYPE[i, j, 0] == 'training' else n_trials_generalization / 2
        BLOC_CENTER_TRIAL[i, j] = -0.5 + previous_index + this_index
        BLOC_STARTING_TRIAL[i, j] = previous_index
BLOC_STARTING_TRIAL
```

### Information about sensory dimensions in array

```python
DIMS = np.full((n_sujs, 2, 2), np.nan, dtype=object)
MODALITY = np.full((n_sujs), np.nan, dtype=object)
for i, suj in enumerate(sujs):
    
    # Get the modality
    MODALITY[i] = DF.loc[DF['id'] == suj, 'modality'].unique()[0]
    
    for j, trial_extra in enumerate([False, True]):

        # Select data
        idx = np.logical_and(DF['id'] == suj, DF['extra'] == trial_extra)
        if DF.loc[idx, 'dim_x'].unique().size == 0: continue
        
        # Put data in array
        DIMS[i, j, 0] = DF.loc[idx, 'dim_x'].unique()[0]
        DIMS[i, j, 1] = DF.loc[idx, 'dim_y'].unique()[0]
```

### Save data as arrays

```python
np.save(dname + 'BLOC_TYPE.npy', BLOC_TYPE)
np.save(dname + 'BLOC_CENTER_TRIAL.npy', BLOC_CENTER_TRIAL)
np.save(dname + 'BLOC_STARTING_TRIAL.npy', BLOC_STARTING_TRIAL)
np.save(dname + 'CORRECT_CATEGORY_BLOC.npy', CORRECT_CATEGORY_BLOC)
np.save(dname + 'RESPONSE_CATEGORY_BLOC.npy', RESPONSE_CATEGORY_BLOC)
np.save(dname + 'CORRECT_BLOC.npy', CORRECT_BLOC)
np.save(dname + 'RT_BLOC.npy', RT_BLOC)
np.save(dname + 'TRIAL_TYPE.npy', TRIAL_TYPE)
np.save(dname + 'DIMS.npy', DIMS)
np.save(dname + 'MODALITY.npy', MODALITY)
np.save(dname + 'SUJS.npy', sujs)
```
