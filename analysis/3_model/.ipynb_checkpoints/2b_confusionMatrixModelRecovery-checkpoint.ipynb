# Model recovery analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.metrics import confusion_matrix
from sklearn.metrics import pairwise_distances
from scipy.stats import pearsonr
```

### Global variables

```python
# Number of trials for model fitting
n_trials = 30

# Number of models to test
model_names = np.array(['random', '1D_x', '1D_y', '2D'])
n_models = model_names.size

# Number of replica of each model
n_sujs = 100

# Number of response labels
labels = np.array(['A', 'B', 'C'])
n_labels = labels.size

# Number of inverse temperature parameter to test 
n_betas = 10
```

```python
betas = np.logspace(*np.log10([0.1, 2]), n_betas)
betas
```

### Compute confusion matrices of models.

```python
# Replace confusion matrix values by idealized CM
CONFUSION_MATRIX_MODELS = np.full((n_models, n_labels, n_labels), np.nan)

# Random
CONFUSION_MATRIX_MODELS[0] = np.array([
    [0.33, 0.33, 0.33],
    [0.33, 0.33, 0.33],
    [0.33, 0.33, 0.33]
])

# 1D_x
CONFUSION_MATRIX_MODELS[1] = np.array([
    [0.5, 0., 0.5],
    [0., 1., 0.],
    [0.5, 0., 0.5]
])

# 1D_y
CONFUSION_MATRIX_MODELS[2] = np.array([
    [0.5, 0.5, 0.],
    [0.5, 0.5, 0.],
    [0., 0., 1.]
])

# 2D
CONFUSION_MATRIX_MODELS[3] = np.array([
    [1., 0., 0.],
    [0., 1., 0.],
    [0., 0., 1.]
])
```

```python
CONFUSION_MATRIX_MODELS.shape
```

```python
# Check confusion matrices 
fig, axs = plt.subplots(n_models, 1, figsize=(1.5, 1.5*n_models), facecolor='w')
labels_plot = ['b', 's', 'h']
model_names_plot = ['RAND', '1Dx', '1Dy', '2D']

for i, (model_name, ax) in enumerate(zip(model_names_plot, axs)):

    # Compute confution matrix (average across participants)
    m = CONFUSION_MATRIX_MODELS[i]

    # Plot matrix
    ax.imshow(m, vmin=0, vmax=1, cmap='viridis')
    ax.set_xticks(range(3))
    ax.set_yticks(range(3))
    ax.set_xticklabels(labels_plot)
    ax.set_yticklabels(labels_plot)
    ax.set_title(model_name, fontweight='bold')

plt.tight_layout()
fig.savefig('figures/fig_1.png')
```

### Generate synthetic data.

```python
CONFUSION_MATRIX_SYNTHETIC = np.zeros((n_models, n_sujs, n_betas, n_labels, n_labels))
```

```python
for i, model in enumerate(model_names):
    for j, suj in enumerate(range(n_sujs)):
        for k, beta in enumerate(betas):
            
            # Generate 30 stimuli
            _stims = np.random.choice(range(n_labels), size=n_trials)
            _stims = np.array(_stims, dtype=np.int)
            
            # Generate 30 responses based on confusion matrix
            _responses = np.zeros(n_trials, dtype=np.int)
            
            for l, _stim in enumerate(_stims):
                
                # Get probability of each response
                _prob = CONFUSION_MATRIX_MODELS[i, _stim]
                
                # Apply softmax with beta inverse temperature
                _prob = np.exp(_prob/beta)/np.sum(np.exp(_prob/beta))
                
                # Sample one response from probabilities
                _responses[l] = np.random.choice(range(n_labels), size=1, p=_prob)

            # Convert responses and stim to labels format (0, 1, 2 --> A, B, C)
            _responses = labels[_responses]
            _stims = labels[_stims]
                
            # Compute confusion matrix
            CONFUSION_MATRIX_SYNTHETIC[i, j, k] = confusion_matrix(
                _stims, _responses, labels=labels, normalize='pred'
            )
```

### Fit the model confusion matrices to the synthetic data

```python
CONFUSION_MATRIX_MODELS.shape
```

```python
CONFUSION_MATRIX_SYNTHETIC.shape
```

```python
CORRELATION_VALUES = np.full((n_models, n_models, n_sujs, n_betas), np.nan)
DISTANCE_VALUES = np.full((n_models, n_models, n_sujs, n_betas), np.nan)

# Loop across model to fit 
for i, model in enumerate(model_names):

    # Loop across every synthetic subject
    for j, model in enumerate(model_names):
        for k, suj in enumerate(range(n_sujs)):
            for l, beta in enumerate(betas):
            
                # Get values
                x = CONFUSION_MATRIX_MODELS[i].flatten()
                y = CONFUSION_MATRIX_SYNTHETIC[j, k, l].flatten()

                # Compute distance value
                DISTANCE_VALUES[i, j, k, l] = pairwise_distances(x[None, :], y[None, :], metric='euclidean')[0, 0]

                # Compute correlation
                if np.unique(x).size == 1 or np.unique(y).size == 1: continue
                CORRELATION_VALUES[i, j, k, l] = pearsonr(x, y)[0]
```

```python
CORRELATION_VALUES[np.isnan(CORRELATION_VALUES)] = 0.
```

### Extract best fitting model for each synthetic subjects.

```python
DISTANCE_VALUES.shape
```

```python
best_model_idxs = np.argmin(DISTANCE_VALUES, axis=0)
#best_model_idxs = np.argmax(CORRELATION_VALUES, axis=0)
BEST_FITTING_MODEL = model_names[best_model_idxs]
```

### Vizualize model recovery.

```python
model_names_plot = ['RAND', '1Dx', '1Dy', '2D']
```

```python
fig, axs = plt.subplots(1, n_betas, figsize=(3*n_betas, 3.2), dpi=300, facecolor='w')

# Loop across betas
for i, (beta, ax) in enumerate(zip(betas, axs)):
    
    # Original model generating synthetic data
    x = np.tile(model_names, (n_sujs, 1)).T.flatten()
    
    # Recovered models
    y = BEST_FITTING_MODEL[:, :, i].flatten()
    
    # Build confusion matrix
    m = confusion_matrix(
        x, y, labels=model_names, #normalize='pred'
    )
    
    # Rotate confusion matrix to align with labels
    m = np.rot90(m, k=1)
    
    # Plot confusion matrix
    ax.imshow(m, aspect='auto')

    # Plot numbers on top of graph
    for j in range(n_models):
        for k in range(n_models):
            ax.text(j, k, str(m[k, j]), va='center', ha='center', color='k')
    
    # Compute performance of a 2D model 
    _prob = np.array([0, 1, 0])
    _prob = np.exp(_prob/beta)/np.sum(np.exp(_prob/beta))
    _perf = _prob[1]

    # Aesthetics
    ax.set_xlabel('Synthetic')
    ax.set_ylabel('Recovered')
    ax.set_title(r'$\beta = $' + str(np.around(beta, 2)) + '\n(2D agent reach ' + str(np.around(_perf*100, 0)) + '%)')
    ax.set_xticks(range(n_models))
    ax.set_xticklabels(model_names_plot)
    ax.set_yticks(range(n_models))
    ax.set_yticklabels(np.flip(model_names_plot))
    
plt.tight_layout()
fig.savefig('figures/fig_6.png')
```
