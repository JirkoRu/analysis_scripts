```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
import pandas as pd
```

### Load data

```python
dname = os.getcwd() + '/../../data/'
DF = pd.read_csv(dname + 'full_data.csv')
```

```python
pd.set_option('display.max_columns', None)
DF.head(2000)
```

### Model random

```python
DF['model_random_prob_A'] = 1./3.
DF['model_random_prob_B'] = 1./3.
DF['model_random_prob_C'] = 1./3.
```

```python
for i in range(len(DF)):
    probs = np.array([
        DF.loc[i, 'model_random_prob_A'],
        DF.loc[i, 'model_random_prob_B'],
        DF.loc[i, 'model_random_prob_C']
    ])
    DF.loc[i, 'model_random_choice'] = np.random.choice(['A', 'B', 'C'], p=probs)
    DF.loc[i, 'model_random_accuracy'] = DF.loc[i, 'model_random_choice'] == DF.loc[i, 'correct_sequence_category']
```

```python
print(DF['model_random_accuracy'].mean()) # Should be 33%
```

### Model 1D_x

```python
for i in range(len(DF)):

    x_seq = np.array([DF.loc[i, 'x0'], DF.loc[i, 'x1'], DF.loc[i, 'x2'], DF.loc[i, 'x3']])
    x_seq_diff = np.abs(np.diff(x_seq))
                
    # All cases
    if np.all(x_seq_diff == np.array([1., 0., 1.])):
        DF.loc[i, 'model_1D_x_prob_A'] = 0.5
        DF.loc[i, 'model_1D_x_prob_B'] = 0.
        DF.loc[i, 'model_1D_x_prob_C'] = 0.5
    elif np.all(x_seq_diff == np.array([0., 1., 0.])):
        DF.loc[i, 'model_1D_x_prob_A'] = 0.5
        DF.loc[i, 'model_1D_x_prob_B'] = 0.
        DF.loc[i, 'model_1D_x_prob_C'] = 0.5
    elif np.all(x_seq_diff == np.array([1., 1., 1.])):
        DF.loc[i, 'model_1D_x_prob_A'] = 0.
        DF.loc[i, 'model_1D_x_prob_B'] = 1.
        DF.loc[i, 'model_1D_x_prob_C'] = 0.
    else:
        raise ValueError()
```

```python
for i in range(len(DF)):
    probs = np.array([
        DF.loc[i, 'model_1D_x_prob_A'],
        DF.loc[i, 'model_1D_x_prob_B'],
        DF.loc[i, 'model_1D_x_prob_C']
    ])
    DF.loc[i, 'model_1D_x_choice'] = np.random.choice(['A', 'B', 'C'], p=probs)
    DF.loc[i, 'model_1D_x_accuracy'] = DF.loc[i, 'model_1D_x_choice'] == DF.loc[i, 'correct_sequence_category']
```

```python
print(DF['model_1D_x_accuracy'].mean()) # Should be 66%
```

### Model 1D_y

```python
for i in range(len(DF)):

    y_seq = np.array([DF.loc[i, 'y0'], DF.loc[i, 'y1'], DF.loc[i, 'y2'], DF.loc[i, 'y3']])
    y_seq_diff = np.abs(np.diff(y_seq))
    
    # All cases
    if np.all(y_seq_diff == np.array([1., 0., 1.])):
        DF.loc[i, 'model_1D_y_prob_A'] = 0.5
        DF.loc[i, 'model_1D_y_prob_B'] = 0.5
        DF.loc[i, 'model_1D_y_prob_C'] = 0.
    elif np.all(y_seq_diff == np.array([0., 1., 0.])):
        DF.loc[i, 'model_1D_y_prob_A'] = 0.5
        DF.loc[i, 'model_1D_y_prob_B'] = 0.5
        DF.loc[i, 'model_1D_y_prob_C'] = 0.
    elif np.all(y_seq_diff == np.array([1., 1., 1.])):
        DF.loc[i, 'model_1D_y_prob_A'] = 0.
        DF.loc[i, 'model_1D_y_prob_B'] = 0.
        DF.loc[i, 'model_1D_y_prob_C'] = 1.
    else:
        raise ValueError()
```

```python
for i in range(len(DF)):
    probs = np.array([
        DF.loc[i, 'model_1D_y_prob_A'],
        DF.loc[i, 'model_1D_y_prob_B'],
        DF.loc[i, 'model_1D_y_prob_C']
    ])
    DF.loc[i, 'model_1D_y_choice'] = np.random.choice(['A', 'B', 'C'], p=probs)
    DF.loc[i, 'model_1D_y_accuracy'] = DF.loc[i, 'model_1D_y_choice'] == DF.loc[i, 'correct_sequence_category']
```

```python
print(DF['model_1D_y_accuracy'].mean()) # Should be 66%
```

### Model 2D

```python
for i in range(len(DF)):

    x_seq = np.array([DF.loc[i, 'x0'], DF.loc[i, 'x1'], DF.loc[i, 'x2'], DF.loc[i, 'x3']])
    x_seq_diff = np.abs(np.diff(x_seq))
    y_seq = np.array([DF.loc[i, 'y0'], DF.loc[i, 'y1'], DF.loc[i, 'y2'], DF.loc[i, 'y3']])
    y_seq_diff = np.abs(np.diff(y_seq))
    shape = np.nan
    if np.all(x_seq_diff == np.array([1., 1., 1.])):
        shape = 'hourglass'
    elif np.all(y_seq_diff == np.array([1., 1., 1.])):
        shape = 'butterfly'
    else:
        shape = 'square'
    
    # All cases
    if shape == 'square':
        DF.loc[i, 'model_2D_prob_A'] = 1.
        DF.loc[i, 'model_2D_prob_B'] = 0.
        DF.loc[i, 'model_2D_prob_C'] = 0.
    elif shape == 'hourglass':
        DF.loc[i, 'model_2D_prob_A'] = 0.
        DF.loc[i, 'model_2D_prob_B'] = 1.
        DF.loc[i, 'model_2D_prob_C'] = 0.
    elif shape == 'butterfly':
        DF.loc[i, 'model_2D_prob_A'] = 0.
        DF.loc[i, 'model_2D_prob_B'] = 0.
        DF.loc[i, 'model_2D_prob_C'] = 1.
    else:
        raise ValueError()
```

```python
for i in range(len(DF)):
    probs = np.array([
        DF.loc[i, 'model_2D_prob_A'],
        DF.loc[i, 'model_2D_prob_B'],
        DF.loc[i, 'model_2D_prob_C']
    ])
    DF.loc[i, 'model_2D_choice'] = np.random.choice(['A', 'B', 'C'], p=probs)
    DF.loc[i, 'model_2D_accuracy'] = DF.loc[i, 'model_2D_choice'] == DF.loc[i, 'correct_sequence_category']
```

```python
print(DF['model_2D_accuracy'].mean()) # Should be 100%
```

### Save model 

```python
DF.to_csv(dname + 'full_data_with_model.csv')
```
