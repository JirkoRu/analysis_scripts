# Plot average performances per bloc

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
```

### Load data

```python
dname = os.getcwd() + '/../../data/'
```

```python
ACCURACY = np.load(dname + 'ACCURACY.npy')
RT = np.load(dname + 'RT.npy')
BLOC_TYPE = np.load(dname + 'BLOC_TYPE.npy', allow_pickle=True)
```

```python
n_sujs, n_blocs, n_trials = ACCURACY.shape
ACCURACY.shape
```

### Plotting parameters

```python
cmap = matplotlib.cm.get_cmap('Set1')
norm = matplotlib.colors.Normalize(vmin=0, vmax=1)
```

```python
bloc_type_colors = {
    'training': cmap(norm(0)),
    'generalization': cmap(norm(.2))
}
```

# Figure 1 - Accuracy

### Compute group average

```python
ACCURACY_M_SUJS = np.nanmean(ACCURACY, axis=2)
ACCURACY_M = np.mean(ACCURACY_M_SUJS, axis=0)
ACCURACY_SE = np.mean(ACCURACY_M_SUJS, axis=0)/np.sqrt(n_sujs)
ACCURACY_M.shape
```

### Plot

```python
# Define figure
fig, ax = plt.subplots(1, 1, figsize=(9, 4), dpi=300, facecolor='w')

# Average per bloc
for i, bloc_type in enumerate(BLOC_TYPE[0]):
    ax.errorbar(
        x = i, 
        y = ACCURACY_M[i],
        yerr = 2 * ACCURACY_SE[i],
        color = bloc_type_colors[bloc_type],
        markersize = 5,
        marker = 'D',
        alpha = 1
    )

# Individual data
for i, bloc_type in enumerate(BLOC_TYPE[0]):
    ax.plot(
        np.ones(n_sujs) * i + np.random.normal(0, 0.05, (n_sujs)), 
        ACCURACY_M_SUJS[:, i],
        color = bloc_type_colors[bloc_type],
        markersize = 3,
        marker = 'o',
        alpha = 0.5, 
        lw = 0
    )

# Aesthetics
ax.set_title('Average accuracy per bloc', fontweight='bold')
ax.set_ylim(-0.05, 1.05)
ax.set_xlabel('Trial #')
ax.set_ylabel('Accuracy')

plt.tight_layout()
fig.savefig('figures/fig_1.png')
```
